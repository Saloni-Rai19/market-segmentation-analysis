```python
import matplotlib.pyplot as plt
from kmodes.kmodes import KModes

cost = []
K = range(1, 10)

for num_clusters in K:
    kmodes = KModes(n_clusters=num_clusters, init='Huang', n_init=5, verbose=1)
    kmodes.fit_predict(df)
    cost.append(kmodes.cost_)

plt.plot(K, cost, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Cost')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
from kmodes.kmodes import KModes
# Select specific columns for clustering
kmodes = KModes(n_clusters=3, init='Huang', n_init=5, verbose=1)
clusters = kmodes.fit_predict(df)

df['Cluster'] = clusters
print(df.head())
```

```python
print("Cluster Modes:")
print(kmodes.cluster_centroids_)
```

```python

import matplotlib.pyplot as plt

sns.countplot(x='Cluster', data=df)
plt.title('Cluster Distribution')
plt.show()
```

```python
import prince
```

```python
# Apply FAMD
famd = prince.FAMD(n_components=3, n_iter = 5, random_state = 101);
famd = famd.fit(df)

# Transform data
famd_components = famd.transform(df)
print(famd_components)

```

```python
plt.scatter(famd_components[0], famd_components[1], alpha=0.5)
plt.xlabel('Component 0')
plt.ylabel('Component 1')
plt.title('Scatter Plot of First two Components')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit_predict(famd_components)

# Add cluster labels to the DataFrame
famd_components['Cluster'] = clusters

# Plotting the clusters
plt.scatter(famd_components[0], famd_components[1], c=clusters, cmap='viridis', alpha=0.5)
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('K-means Clustering on FAMD Components')
plt.show()
```

```python
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
plt.scatter(famd_components[0], famd_components[1], famd_components[2], alpha=0.5)
plt.xlabel('Component 0')
plt.ylabel('Component 1')
plt.title('Scatter Plot of First three Components')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(famd_components[0], famd_components[1], famd_components[2], alpha=0.5)
ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')
ax.set_zlabel('Component 3')
ax.set_title('3D Scatter Plot of First Three Components')
plt.show()
```

```python
import seaborn as sns
from kmodes.kmodes import KModes
import prince
# Calculate segment profiles
segment_profiles = df.groupby('Cluster').mean().reset_index()

# Melt the DataFrame
segment_profiles_melted = segment_profiles.melt(id_vars='Cluster', var_name='variable', value_name='value')

# Plot the segment profile plot
plt.figure(figsize=(12, 8))
sns.lineplot(data=segment_profiles_melted, x='variable', y='value', hue='Cluster', marker='o')
plt.title('Segment Profile Plot')
plt.xlabel('Variable')
plt.ylabel('Mean Value')
plt.legend(title='Cluster')
plt.xticks(rotation=45)
plt.show()
```
